**Design Implementation**

The implementation used follows the Uniform Cost Seach algorithm as well as incorporating my own challenges relevant to my scenario.

```python
import heapq

#instantiating the weighted graph
banks = {
    "A": {"B":{"cost":4,"time":5}, "C":{"cost":3,"time":4}, "D":{"cost":7,"time":8}},
    "B": {"A":{"cost":4,"time":4}, "E":{"cost":5,"time":6}, "F":{"cost":6,"time":7}},
    "C": {"A":{"cost":3,"time":4}, "D":{"cost":3,"time":3}, "F":{"cost":4,"time":5}, "G":{"cost":6,"time":7}},
    "D": {"A":{"cost":7,"time":8}, "C":{"cost":3,"time":3}, "G":{"cost":3,"time":4}},
    "E": {"B":{"cost":5,"time":6}, "I":{"cost":4,"time":5}, "J":{"cost":7,"time":8}},
    "F": {"B":{"cost":6,"time":7}, "C":{"cost":4,"time":5}, "G":{"cost":2,"time":2}, "J":{"cost":3,"time":4}, "K":{"cost":6,"time":7}},
    "G": {"C":{"cost":6,"time":7}, "D":{"cost":3,"time":4}, "F":{"cost":2,"time":2}, "L":{"cost":4,"time":5}},
    "H": {"D":{"cost":4,"time":5}, "M":{"cost":5,"time":6}, "L":{"cost":6,"time":7}},
    "I": {"E":{"cost":4,"time":5}, "N":{"cost":3,"time":4}},
    "J": {"E":{"cost":7,"time":8}, "F":{"cost":3,"time":4}, "N":{"cost":6,"time":7}, "O":{"cost":4,"time":5}},
    "K": {"F":{"cost":6,"time":7}, "O":{"cost":3,"time":3}, "P":{"cost":5,"time":6}},
    "L": {"G":{"cost":4,"time":5}, "H":{"cost":6,"time":7}, "P":{"cost":4,"time":5}, "Q":{"cost":6,"time":7}},
    "M": {"H":{"cost":5,"time":6}, "Q":{"cost":3,"time":3}, "R":{"cost":5,"time":6}},
    "N": {"I":{"cost":3,"time":4}, "J":{"cost":6,"time":7}, "S":{"cost":4,"time":5}},
    "O": {"J":{"cost":4,"time":5}, "K":{"cost":3,"time":3}, "P":{"cost":2,"time":2}, "S":{"cost":3,"time":3}},
    "P": {"K":{"cost":5,"time":6}, "L":{"cost":4,"time":5}, "O":{"cost":2,"time":2}, "T":{"cost":5,"time":6}},
    "Q": {"L":{"cost":6,"time":7}, "M":{"cost":3,"time":3}, "T":{"cost":4,"time":5}},
    "R": {"M":{"cost":5,"time":6}, "T":{"cost":3,"time":3}},
    "S": {"N":{"cost":4,"time":5}, "O":{"cost":3,"time":3}, "T":{"cost":3,"time":3}},
    "T": {"P":{"cost":4,"time":6}, "Q":{"cost":4,"time":5}, "R":{"cost":3,"time":3}, "S":{"cost":3,"time":3}}
}

#instantiating the high-risk nodes and their penalty
high_risk_nodes = {"F":7,"H":6,"J":6,"M":5,"I":7,"L":8}

max_budget = 30

def uniform_cost_search(starting_node,target_node,graph):
    #frontier to be stored by increasing cumulative cost
    #stored as tuples of (cost,current_node,current path)
    #we initialise the frontier with a cumulative cost of 0,
    #current node as the starting node, and the current path starting from the starting node
    frontier = [(0,starting_node,[starting_node])]
    visited = set()
    #used to store the cheapest cost of finding each node
    cost_so_far = {starting_node:0}

    alpha = 0.7
    beta = 0.3

    while frontier:
        #at the beginning of every iteration we want to sort the frontier by cumulative cost
        heapq.heapify(frontier)
        (score,node,path)=heapq.heappop(frontier)

        if node == target_node:
            print(f"Final Score {score}")
            return "Optimal Path", path

        else:
            if node in visited:
                continue
            else:
                visited.add(node)

            #iterates through the neighbours of the current node and adds them to the frontier
            for neighbour in graph[node].keys():

                if neighbour in high_risk_nodes:
                    #the penalty of traversing through a high-risk node isn't adjusted by an alpha and beta value because
                    #the risk is applied independently
                    new_score = score + (alpha * (graph[node][neighbour]["cost"])) + (beta * (graph[node][neighbour]["time"])) + high_risk_nodes[neighbour]
                else:
                    new_score = score + (alpha * (graph[node][neighbour]["cost"])) + (beta * (graph[node][neighbour]["time"]))

                #prevents neighbours being added to the path if the budget is exceeded
                if new_score > max_budget:
                    continue

                if neighbour not in cost_so_far or new_score < cost_so_far[neighbour]:
                    cost_so_far[neighbour] = new_score
                    heapq.heappush(frontier, (new_score, neighbour, path + [neighbour]))

    #output when there is no valid path which meets the budget requirements
    print("No valid path which meets the maximum budget")
    print(f"The closest path is {path} with a cost of {new_score}")
```

```python
print("Test Case 1: Normal Budget of 30 ")
uniform_cost_search("A","T",banks)

#print("Test Case 2: Low budget of 25")
#max_budget = 25
#uniform_cost_search("A","T",banks)
```
