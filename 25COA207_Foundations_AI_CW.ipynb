```python
import heapq
banks = {
    "A": {"B":4, "C":3, "D":7},
    "B": {"A":4, "E":5, "F":6},
    "C": {"A":3, "D":3, "F":4, "G":6},
    "D": {"A":7, "C":3, "G":3},
    "E": {"B":5, "I":4, "J":7},
    "F": {"B":6, "C":4, "G":2, "J":3, "K":6},
    "G": {"C":6, "D":3, "F":2, "L":4},
    "H": {"D":5, "M":5, "L":6},
    "I": {"E":4, "N":3},
    "J": {"E":7, "F":3, "N":6, "O":4},
    "K": {"F":6, "O":3, "P":5},
    "L": {"G":4, "H":6, "P":4, "Q":6},
    "M": {"H":5, "Q":3, "R":5},
    "N": {"I":3, "J":6, "S":4},
    "O": {"J":4, "K":3, "P":2, "S":3},
    "P": {"K":5, "L":4, "O":2, "T":5},
    "Q": {"L":6, "M":3, "T":4},
    "R": {"M":5, "T":3},
    "S": {"N":4, "O":3, "T":3},
    "T": {"P":5, "Q":4, "R":3, "S":3}
}

def uniform_cost_search(starting_node,target_node,graph):
    #priority queue to be stored by increasing cumulative cost
    #stored as tuples of (cost,current_node,current path)
    frontier = [(0,starting_node,[starting_node])]

    explored = set()
    while frontier:
        heapq.heapify(frontier)
        (cost,node,path)=heapq.heappop(frontier)

        if node == target_node:
            print(f"cost {cost}")
            print(f"explored set {explored}")
            return path

        else:
            if node in explored:
                continue
            else:
                explored.add(node)

            #iterates through the neighbours of the current node and adds them to the frontier
            for neighbour,weight in graph[node].items():
                new_cost = cost + weight
                if neighbour not in frontier:
                    heapq.heappush(frontier,(new_cost,neighbour,path+[neighbour]))

                #replaces the neighbour in frontier if we found a shorter path
                elif neighbour in frontier and new_cost<frontier[neighbour]:
                    frontier[neighbour] = new_cost

    return None

starting_node = "A"
target_node = "T"
uniform_cost_search(starting_node,target_node,banks)
```
